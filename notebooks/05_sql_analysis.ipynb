```python
# Import libraries
import pandas as pd
import sqlite3
```

```python
# Connect to SQLite database
conn = sqlite3.connect(
    "../data/books.db"
)
```

```python
query = """
SELECT *
FROM books
LIMIT 5;
"""

pd.read_sql_query(query,conn)
```

```python
# Summary of missing data in the books dataset
query = """
SELECT
    COUNT(*) AS total_books,
    SUM(
        CASE
            WHEN category IS NULL
            THEN 1
            ELSE 0
        END
    ) AS missing_categories,
    SUM(
        CASE
            WHEN stock_quantity IS NULL
            THEN 1
            ELSE 0
        END
    ) AS missing_stock
FROM books;
"""
pd.read_sql_query(query,conn)
```

```python
# Books by category
query = """
SELECT
    category,
    COUNT(*) AS total_books
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY total_books DESC;
"""

pd.read_sql_query(query,conn)
```

```python
# Total number of books by category
query = """
SELECT
    category,
    COUNT(*) AS total_books
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY total_books DESC
LIMIT 10
"""

pd.read_sql_query(query,conn)
```

```python
# Top 10 most expensive books
query = """
SELECT
    title,
    price
FROM books
ORDER BY price DESC
LIMIT 10
"""

pd.read_sql_query(query,conn)
```

```python
# Average price of books by category
query = """
SELECT
    category,
    ROUND(
        AVG(price),
        2
    ) AS average_price
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY average_price DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Average rating of books by category
query = """
SELECT
    category,
    ROUND(
        AVG(rating),
        1
    ) AS average_rating
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY average_rating DESC
LIMIT 10
"""

pd.read_sql_query(query,conn)
```

```python
# Books with 5-star rating
query = """
SELECT
    title,
    category,
    price
FROM books
WHERE category IS NOT NULL
AND category <> ''
AND rating = 5
ORDER BY price DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Total number of 5-star books by category
query = """
SELECT
    category,
    COUNT(*) AS five_star_books
FROM books
WHERE category IS NOT NULL
AND category <> ''
AND rating = 5
GROUP BY category
ORDER BY five_star_books DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Total stock quantity of books by category
query = """
SELECT
    category,
    SUM(stock_quantity) AS total_stock
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY total_stock DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Top 10 books with the highest stock quantity
query = """
SELECT
    title,
    category,
    stock_quantity
FROM books
WHERE category IS NOT NULL
AND category <> ''
ORDER BY stock_quantity DESC
LIMIT 10
"""

pd.read_sql_query(query,conn)
```

```python
# Total inventory value by category
query = """
SELECT
    category,
    ROUND(
        SUM(price * stock_quantity),
        2
    ) AS inventory_value
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY inventory_value DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Average price of books by rating
query = """
SELECT
    rating,
    ROUND(
        AVG(price),
        2
    ) AS average_price
FROM books
GROUP BY rating
ORDER BY rating
"""

pd.read_sql_query(query,conn)
```

```python
# Minimum and maximum price of books by category
query = """
SELECT
    category,
    MIN(price) AS min_price,
    MAX(price) AS max_price
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY max_price DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Average stock quantity of books by category
query = """
SELECT
    category,
    ROUND(
        AVG(stock_quantity),
        1
    ) AS average_stock
FROM books
WHERE category IS NOT NULL
AND category <> ''
GROUP BY category
ORDER BY average_stock DESC
"""

pd.read_sql_query(query,conn)
```

```python
# Summary statistics of the books dataset
query = """
SELECT
    COUNT(*) AS total_books,
    COUNT(DISTINCT category) AS total_categories,
    ROUND(AVG(price),2) AS average_price,
    ROUND(AVG(rating),2) AS average_rating
FROM books
WHERE category IS NOT NULL
AND category <> ''
"""

pd.read_sql_query(query,conn)
```

```python
# Average price and rating of books by category (only categories with at least 10 books)
query = """
SELECT
    category,
    COUNT(*) AS books,
    ROUND(AVG(price),2) AS avg_price,
    ROUND(AVG(rating),2) AS avg_rating
FROM books
WHERE category IS NOT NULL
GROUP BY category
HAVING COUNT(*) >= 10
ORDER BY avg_rating DESC;
"""
pd.read_sql_query(query,conn)
```
