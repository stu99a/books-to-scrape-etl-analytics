```python
# Import libraries
import pandas as pd
import sqlite3
```

```python
# Load cleaned dataset
books_df = pd.read_csv(
    "../data/processed/books_cleaned.csv"
)

books_df.head()
```

```python
# Create database connection
conn = sqlite3.connect(
    "../data/books.db"
)
```

```python
# Load data into SQLite
books_df.to_sql(
    "books",
    conn,
    if_exists="replace",
    index=False
)
```

```python
# Verify table creation
query = """
SELECT *
FROM books
LIMIT 5
"""

pd.read_sql_query(query,conn)
```

```python
# Check row count
query = """
SELECT COUNT(*) AS total_books
FROM books
"""

pd.read_sql_query(query,conn)
```

```python
# Close connection
conn.close()
```
